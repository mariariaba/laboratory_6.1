```python
import random

min = -100
max = 100

random_number = random.randint(min, max)

print(f"Ваше випадкове число: {random_number}")

if random_number > 0: print("Це додатнє число")
if random_number < 0: print("Це від'ємне число")
```

# **Пояснення коду** №1:


1. Встановлюємо діапазон для генерації випадкового числа;
2. Генеруємо випадкове число у вказаному діапазоні;
3. Визначаємо чи вона є більшим/меншим від 0;
4. Виводимо випадкове число;
5. Виводимо фразу "Це число додатнє (від'ємне)" відповідно до пункту 3.


Цей код використовує *'random'* щоб генерувати випадкове ціле число між *'min'* та *'max'* відповідно до Ваших потреб. Також тут я використала команту *'if'* щоб здійснити перевірку числа (чи є воно додатнім чи від'ємним) щоб вивести фразу відповідно до перевірки.


















```python
import random

random_numbers = [random.randint(1, 100) for _ in range(10)]
print("Початковий список:", random_numbers)

random_numbers.sort()
print("Відсортований список:", random_numbers)
```

# Пояснення коду №2:




1.   Генеруємо список випадкових чисел;
2.   Виводимо початковий список;
3.   Сортуємо список;
4.   Виводимо відсортований список.




Ця програма використовує модуль *'random'*, щоб згенерувати список випадкових чисел в діапазоні від 1 до 100, потім сортує їх і виводить обидва списки (*початковий і відсортований*). Ви можете запустити цей код у середовищі **Python**, щоб побачити результат.
